```python
res = {}
extr_data = './data/test/extraction.csv'
man_data = './data/test/manual.csv'
df_ext = pd.read_csv(extr_data)
df_ext = shuffle(df_ext)
df_ext[:10]
```

```python
df_man = pd.read_csv(man_data)
df_man = shuffle(df_man)
df_man[:10]
```

```python
df_ext = df_ext.sort_values(by=['time'], ascending=False)
df_man = df_man.sort_values(by=['time'], ascending=False)

std_ext = round(np.std(df_ext['time']), 2)
mean_ext = round(np.mean(df_ext['time']), 2)
std_man  =  round(np.std(df_man['time']), 2)
mean_man  =  round(np.mean(df_man['time']), 2)

print('duration extraction: mean={m} std={std}'.format(m=mean_ext, std=std_ext))
print('duration manual: mean={m} std={std}'.format(m=mean_man, std=std_man ))
```

```python
relative_error_percentages = {}
absolute_error_percentages = {}
amount_fields = 40
amount_tests = 9

sum_fields = amount_fields * amount_tests
total = sum(sum_errors.values())
for label, amount in sum_errors.items():
    relative_error_percentages[label] = (total/100) * amount
    
for label, amount in sum_errors.items():
    absolute_error_percentages[label] = round(100*amount/sum_fields, 2)

print('{total} errors were registered, which are {each} errors each process. The percentage error rates per field are:\n'.format(total=total, each=total/40))


for label, amount in sum_errors.items():
     print('An absolute error rate of {amount}% in extracting data for "{label}".'.format(amount=round(100*amount/sum_fields, 2), label=label))
print('An average absolute error rate of {mean}%.'.format(mean=round(np.mean(list(absolute_error_percentages.values())),2)))
print('\n')
for label, perc in error_percentages.items():
    print('An relative error rate of {amount}% in extracting data for "{label}".'.format(amount=perc, label=label))

print('An average relative error rate of {amount}%'.format(amount=round(np.mean(list(error_percentages.values())), 2)))

print('\nIn an amount of {sum_field} has been {amount_errors} errors registered which corresponds to {perc}% errors.'
      .format(sum_field=sum_fields, amount_errors=total, perc=round(100*(total/sum_fields),2)))


x = np.arange(len(list(absolute_error_percentages.values())))
width = 0.35  

fig, ax = plt.subplots()
fig.set_size_inches(10.5, 6.5)
ax.bar(x - width/2, absolute_error_percentages.values(), width,label='error rates')
ax.bar(x + width/2, round(np.mean(list(absolute_error_percentages.values())), 2), width, label='average')
ax.set_ylabel('Percentage')
ax.set_title('Error rates')
ax.set_xticks(x)
ax.set_xticklabels((list(absolute_error_percentages.keys())))
ax.legend()

fig.tight_layout()

plt.show()    
```

```python
class NumpyEncoder(json.JSONEncoder):
    """ Special json encoder for numpy types """
    def default(self, obj):
        if isinstance(obj, (np.int_, np.intc, np.intp, np.int8,
            np.int16, np.int32, np.int64, np.uint8,
            np.uint16, np.uint32, np.uint64)):
            return int(obj)
        elif isinstance(obj, (np.float_, np.float16, np.float32, 
            np.float64)):
            return float(obj)
        elif isinstance(obj,(np.ndarray,)): #### This is the fix
            return obj.tolist()
        return json.JSONEncoder.default(self, obj)
```
